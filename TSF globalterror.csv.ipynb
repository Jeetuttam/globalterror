# We don't need all the columns to work with so we either drop unnecessry columns or create a new datafram

```python
#let's create a new datafram 
terror_df = global_terr[['year','month','day','Country','state','Region','city','latitude','longitude','AttackType','Killed','wounded','target','summary','group','Target_type','weapontype','motive']]
terror_df.head()
```

```python
terror_df.info() # here is the information of new data set that created using new var terror_df
```

```python
terror_df.describe(include='all') # describing the data for all the columns
```

# checking Null value

```python
terror_df.isna().sum()
```

# cleaning the data

```python
terror_df['Killed']=terror_df['Killed'].fillna(0).astype(int)
terror_df['state']=terror_df['state'].fillna(0).astype(object)
terror_df['city']=terror_df['city'].fillna(0).astype(object)
terror_df['target']=terror_df['target'].fillna(0)
```

```python
terror_df['wounded']=terror_df['wounded'].fillna(0).astype(int)
```

__Most Attacktype are Bombing/Explosion__

```python
## lets create a new var to check how many people are killes by which type
```

```python
attack_killed = terror_df[['AttackType','Killed']].groupby(["AttackType"],axis=0).sum()
attack_killed
```

## here we see most of people were killed by Armed Assault

```python
plt.subplots(figsize=(12,6)) # check the plot which will clear you most
sns.barplot(attack_killed.index, attack_killed.Killed.values,palette="summer_r")
plt.title('Number of People Killed in each attack type')
plt.xlabel('Types ot attacks')
plt.ylabel('Number of people killed')
plt.xticks(rotation= 45)
plt.show()
```
